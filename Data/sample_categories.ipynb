```python
#detailed category
category = []
for i in range(len(ls)):
    category.append(ls[i][1])
```

```python
res = {genes[i]: category[i] for i in range(len(genes))}
```

```python
res
```

```python
C = (pd.Series(res))
```

```python
df_category = pd.Series.to_frame(C)
```
